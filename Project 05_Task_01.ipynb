Task 1.1 :- Overview of the users behavior

```python
# Aggregating the required information per user

# Grouping data by 'MSISDN/Number' which represents each user
user_group = df.groupby('MSISDN/Number')

# Aggregating the required information
aggregated_data = user_group.agg(
    number_of_xDR_sessions = pd.NamedAgg(column='Bearer Id', aggfunc='count'),
    total_session_duration = pd.NamedAgg(column='Dur. (ms)', aggfunc='sum'),
    total_DL_data = pd.NamedAgg(column='Total DL (Bytes)', aggfunc='sum'),
    total_UL_data = pd.NamedAgg(column='Total UL (Bytes)', aggfunc='sum'),
    social_media_DL = pd.NamedAgg(column='Social Media DL (Bytes)', aggfunc='sum'),
    social_media_UL = pd.NamedAgg(column='Social Media UL (Bytes)', aggfunc='sum'),
    google_DL = pd.NamedAgg(column='Google DL (Bytes)', aggfunc='sum'),
    google_UL = pd.NamedAgg(column='Google UL (Bytes)', aggfunc='sum'),
    email_DL = pd.NamedAgg(column='Email DL (Bytes)', aggfunc='sum'),
    email_UL = pd.NamedAgg(column='Email UL (Bytes)', aggfunc='sum'),
    youtube_DL = pd.NamedAgg(column='Youtube DL (Bytes)', aggfunc='sum'),
    youtube_UL = pd.NamedAgg(column='Youtube UL (Bytes)', aggfunc='sum'),
    netflix_DL = pd.NamedAgg(column='Netflix DL (Bytes)', aggfunc='sum'),
    netflix_UL = pd.NamedAgg(column='Netflix UL (Bytes)', aggfunc='sum'),
    gaming_DL = pd.NamedAgg(column='Gaming DL (Bytes)', aggfunc='sum'),
    gaming_UL = pd.NamedAgg(column='Gaming UL (Bytes)', aggfunc='sum'),
    other_DL = pd.NamedAgg(column='Other DL (Bytes)', aggfunc='sum'),
    other_UL = pd.NamedAgg(column='Other UL (Bytes)', aggfunc='sum')
)

# Calculating the total data volume for each application
aggregated_data['total_social_media_data'] = aggregated_data['social_media_DL'] + aggregated_data['social_media_UL']
aggregated_data['total_google_data'] = aggregated_data['google_DL'] + aggregated_data['google_UL']
aggregated_data['total_email_data'] = aggregated_data['email_DL'] + aggregated_data['email_UL']
aggregated_data['total_youtube_data'] = aggregated_data['youtube_DL'] + aggregated_data['youtube_UL']
aggregated_data['total_netflix_data'] = aggregated_data['netflix_DL'] + aggregated_data['netflix_UL']
aggregated_data['total_gaming_data'] = aggregated_data['gaming_DL'] + aggregated_data['gaming_UL']
aggregated_data['total_other_data'] = aggregated_data['other_DL'] + aggregated_data['other_UL']

# Dropping intermediate columns
aggregated_data.drop(columns=['social_media_DL', 'social_media_UL', 
                              'google_DL', 'google_UL', 'email_DL', 
                              'email_UL', 'youtube_DL',  'youtube_UL',
                              'netflix_DL', 'netflix_UL', 'gaming_DL', 'gaming_UL',
                              'other_DL', 'other_UL'], inplace=True)

aggregated_data.head()
    
```

```python
# Identify Missing Values

null_values = df.isnull().sum()
```

There are many missing values in the dataset

```python
# Percentage % of missing values

columns_with_na = (df.isnull().sum() / len(df)) * 100

# Filter out the features and sort

columns_with_na = columns_with_na[columns_with_na > 0].sort_values(ascending=False)

# Print the feature name and the percentage of missing values

print(columns_with_na)
```

Here, There are many features which have more than 50 % missing values. We can consider dropping these values

```python
# Dropping the columns which have more than 50% missing values

columns_to_drop  = df.columns[df.isnull().mean() > 0.5]

data_cleaned = df.drop(columns=columns_to_drop)

# Displaying the information of the cleaned Dataframe
data_cleaned.info()
```

```python
# Replacing missing values with the mean of corresponding column

# Calculate the mean only for numerical columns
numerical_means = data_cleaned.select_dtypes(include=['float64', 'int64']).mean()

# Fill missing values in numerical columns with their respective means
df_filled = data_cleaned.fillna(numerical_means)

df_filled.isnull().sum()
```

```python
# Treating missing values of categorical data with the Mode of corresponding column

# Replace missing values in the column 'Last Location Name' with its mode
mode_of_lastlocation = df_filled['Last Location Name'].mode()[0]
df_filled['Last Location Name'].fillna(mode_of_lastlocation, inplace=True)

# Replace missing values in the column 'Last Location Name' with its mode
mode_of_handset_man = df_filled['Handset Manufacturer'].mode()[0]
df_filled['Handset Manufacturer'].fillna(mode_of_handset_man, inplace=True)

# Replace missing values in the column 'Handset Type' with its mode
mode_of_handset_type = df_filled['Handset Type'].mode()[0]
df_filled['Handset Type'].fillna(mode_of_handset_type, inplace=True)

df_filled.isnull().sum()
```

The missing values in the dataset have been successfully identified and replaced

```python
# Outlier Detection and Treatment

# Function to detect and treat outliers
def replace_outliers_with_mean(dataframe):
    # Select only numeric columns (exclude datetime and non-numeric columns)
    numeric_cols = dataframe.select_dtypes(include=[np.number]).columns
    numeric_df = dataframe[numeric_cols].copy()

    # Calculate IQR for numeric columns
    Q1 = numeric_df.quantile(0.25)
    Q3 = numeric_df.quantile(0.75)
    IQR = Q3 - Q1

    # Identify outliers
    is_outlier = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR)))

    # Replace outliers with NaN
    numeric_df[is_outlier] = np.nan

    # Replace NaN with column mean
    for column in numeric_df.columns:
        numeric_df[column].fillna(numeric_df[column].mean(), inplace=True)
    
    # Update the original dataframe
    dataframe[numeric_cols] = numeric_df

    return dataframe

# Treating outliers in the dataset
data_treated = replace_outliers_with_mean(df_filled)

# Displaying the summary statistics of the treated dataset to verify the treatment
data_treated.describe()
```

```python
# Bivariate Analysis :- In this case, we will examine the relationship between the data usage of each application and the total data usage

# A suitable method for this analysis is to use correlation coefficients

# Calculating the total data usage (DL + UL) for each application
aggregated_data['total_data'] = aggregated_data['total_DL_data'] + aggregated_data['total_UL_data']

# Calculating Pearson's correlation coefficients between each applications's data usage and total data usage
correlation_coefficients = {
    'Social Media' : aggregated_data['total_social_media_data'].corr(aggregated_data['total_data']),
    'Google' : aggregated_data['total_google_data'].corr(aggregated_data['total_data']),
    'Email' : aggregated_data['total_email_data'].corr(aggregated_data['total_data']),
    'Youtube' : aggregated_data['total_youtube_data'].corr(aggregated_data['total_data']),
    'Netflix' : aggregated_data['total_netflix_data']. corr(aggregated_data['total_data']),
    'Gaming' : aggregated_data['total_gaming_data'].corr(aggregated_data['total_data']),
    'Other' : aggregated_data['total_other_data'].corr(aggregated_data['total_data'])
}

correlation_coefficients_df = pd.DataFrame.from_dict(correlation_coefficients, orient='index', columns=['Correlation with Total Data'])
correlation_coefficients_df.sort_values(by='Correlation with Total Data', ascending=False)
```

Interpretations :-

Gaming : With a correlation coefficient very close to 1, there is a very strong positive correlation between gaming data usage and total data usage. This implies that users who consume more data on gaming tend to have higher overall data usage. gaming appears to be a significant contributor to total data consumption.

Youtube, Netflix, Google and Email : These applications show moderate positive correlations with total data usage. While they contribute to the total data usage, their impact is not as pronounced as gaming. Users who consume more data on these applications tend to have moderately higher overall data usage.

Social Media and Other : The correlations are positive but relatively weaker compared to other applications. This suggests that while there is some relationship between data usage on these applications and total data usage, it is less significant.

```python
# Variable Transformations :- Segment the users into the top five decile classes based on the total duration for all sessions
# Compute the total data (DL+UL) per decile class

# Rank users based on total session duration
aggregated_data['session_duration_rank'] = aggregated_data['total_session_duration'].rank(pct=True)

# Segment these rankings into deciles
aggregated_data['decile_class'] = pd.qcut(aggregated_data['session_duration_rank'], 10, labels=False) + 1

# Focusing on the top five deciles (50th to 100th percentile)
top_five_deciles = aggregated_data[aggregated_data['decile_class'] > 5]

# Calculate the Total Data (DL + UL) for each of these decile classes
total_data_per_decile = top_five_deciles.groupby('decile_class')['total_data'].sum()

total_data_per_decile


```

Observations :-

Increasing Data Usage in Higher Deciles : There's a general increase in total data usage as we move to higher deciles. Particularly, the 10th decile, representing users with the longest session durations, Shows significantly higher data usage compared to other deciles.

```python
# Dimensionality Reduction :– We will perform a principal component analysis to reduce the dimensions of our data 

# Let's perform PCA on our dataset, specifically focusing on the application data (Social Media, Google, Email, Youtube, Netflix, Gaming and Other data)

# Import the necessary packages
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Selecting the relevant features for PCA
features = ['total_social_media_data', 'total_google_data', 'total_email_data', 'total_youtube_data', 'total_netflix_data', 
            'total_gaming_data', 'total_other_data']

# Standardizing the features
x = aggregated_data.loc[:, features].values
x = StandardScaler().fit_transform(x)

# Performing PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x)

# Creating a dataframe with principal components
principal_df = pd.DataFrame(data = principal_components, columns = ['principal component_1', 'principal component_2'])

# Variance explained by each component
explained_variance = pca.explained_variance_ratio_

explained_variance, principal_df.head()

```

Interpretations :-

1) Explained Variance by Components :
- The array [0.70735526, 0.05912166] represents the proportion of the dataset's variance that each principal component accounts for.
- The first principal component accounts for approximately 70.74% of the variance in the data.
- The second principal component accounts for about 5.91% of the variance.
- Together, these two components explain about 76.65% of the total variance in the dataset.

2) Principal Component Scores :
- The table shows the scores of the first five observations on the two principal components
- The first principal component represents a general measure of data usage across all applications, as it captures the most variance.
- The second principal component might capture contrasting usage patterns among the applications.

3) Data Compression :  By reducing the dataset to these two components, we have significantly compressed the data while still retaining a majority of the information.

4) Application for Network operators : It might be particularly useful for segmenting users or understanding overall usage trends in a more streamlined manner.
